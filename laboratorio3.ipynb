```python
#Nodos.py
class Nodo:
    def __init__(self, estado, hijo=None):
        self.estado = estado #posibles movimientos, cordenadas, valores en relaion a un problema
        self.hijo = None 
        self.padre = None
        self.accion = None
        self.acciones = None #conjunto de acciones 
        self.costo = None
        self.set_hijo(hijo) #

    def set_estado(self, estado):
        self.estado = estado

    def get_estado(self):
        return self.estado

    def set_hijo(self, hijo):
        self.hijo = hijo
        if self.hijo is not None:
            for s in self.hijo:
                s.padre = self

    def get_hijo(self):
        return self.hijo
    
    def set_padre(self, padre):
        self.padre = padre

    def get_padre(self):
        return self.padre
    
    def set_accion(self, accion):
        self.padre = accion

    def get_accion(self):
        return self.accion

    def set_acciones(self, acciones):
        self.acciones = acciones

    def get_acciones(self):
        return self.acciones

    def set_costo(self, costo):
        self.costo = costo

    def get_costo(self):
        return self.costo

    def equal(self, Nodo): # compraracion de nodos
        if self.get_estado() == Nodo.get_estado():
            return True
        else:
            return False

    def en_lista(self, lista_nodos): # si el nodo ya existe basado en el estado
        enlistado = False
        for n in lista_nodos:
            if self.equal(n):
                enlistado = True
        return enlistado

    def __str__(self):
        return str(self.get_estado())
```

```python
def intercambio(pos1, pos2, list):
  list[pos1], list[pos2] = list[pos2], list[pos1] 
  return list
```

```python
def busqueda_BPA_solucion(estado_inicial, solucion):
    resuelto = False
    nodos_visitados = [] # ya visitados
    nodos_frontera = [] # para visitar
    hijo = []
    
    nodo_raiz = Nodo(estado_inicial)
    if nodo_raiz == solucion:
      return nodo_raiz
    nodos_frontera.append(nodo_raiz) #cracion de cola
    while (not resuelto) and len(nodos_frontera) != 0:
        nodo_actual = nodos_frontera.pop(0) #sacamos el primer elemento de nodos_frontera
        # extraer nodo y añadirlo a visitados
        nodos_visitados.append(nodo_actual) # añadimos nodo_Actual a nodos_visitados para no repetir
        if nodo_actual.get_estado() == solucion:
            # solucion encontrada
            resuelto = True
            return nodo_actual
        else:
            estado_nodo = nodo_actual.get_estado() # expandir el nodo hijo, y como esta conformado
            posiciones = []
            for i in range(len(solucion)-1): # desplegamos 9 conbinaciones
              hijo = []
              for j in range(len(solucion)): # 10
                hijo.append(estado_nodo[j]) # [estado_nodo[0], estado_nodo[1], estado_nodo[2], estado_nodo[3]]
              # print(hijo)
              cabio_posiciones = intercambio(i, i+1, hijo)
              # print(cabio_posiciones) #bien 
              nodos_hijos = Nodo(cabio_posiciones)
              x = posiciones.append([nodos_hijos]) # [hijo_izquierda, hijo_centro, hijo_derecha]
              # print(posiciones)
              if not nodos_hijos.en_lista(nodos_visitados) and not nodos_hijos.en_lista(nodos_frontera):
                nodos_frontera.append(nodos_hijos)        
            nodo_actual.set_hijo(x)
          
if __name__ == "__main__":
    estado_inicial = [6, 4, 0, 1, 3, 5, 2, 7]
                    #[6, 4, 0, 1, 3, 5, 2] demora 12 seg. en hallar la solucion 
                    #[6, 4, 0, 1, 3, 5, 2, 7] demora 3min 29seg. en hallar la solucion  
                    #[6, 4, 0, 1, 3, 5, 2, 8, 7] demora demasiado en hallar la solucion
    solucion = [0, 1, 2, 3, 4, 5, 6, 7]

#COLAS -> FIFO (first in, first out), el primero en entrar es el primero en salir.
#PILAS -> LIFO (last in, first out), el último que entra es el primero en salir.
    nodo_solucion = busqueda_BPA_solucion(estado_inicial, solucion)
    # mostrar resultado
    resultado = []
    nodo_actual = nodo_solucion
    print('Solucion:', nodo_solucion)
    while nodo_actual.get_padre() is  not None:
        print('resul', resultado)
        resultado.append(nodo_actual.get_estado())
        nodo_actual = nodo_actual.get_padre()

    resultado.append(estado_inicial)
    resultado.reverse()
    print(resultado)
```
